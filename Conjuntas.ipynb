```python
#Función que calcula las probabilidades de la matriz F
#i es el valor de spam, j la Wi que se está checando y m la matriz de valores de probabilidades de wn dado spam=1/0
def probabilidad(i,j,m):
    
    W=Ws[j]
    V=Vs[j]
    resnum=1
    resden=1
    resultadonum=0
    resultadoden=0
    
    if i == 1: #s es lo que define en qué parte de la matriz está, en spam=0 o spam=1
        spam=0.75
        s=3
    else:
        spam=0.25
        s=1
        
    #print("V={}".format(V))
    #print("len(V)={}".format(len(V)))
    for a in range(2**len(V)): #Sacamos el numerador
        b=bin(a)
        d=list(b)
        for c in range(len(d)):
            if d[c]=="0":
                d[c]="1"
            else:
                d[c]="0"
        b=list(d)
        for num in range(len(b)):
            b[num]=int(b[num])
        #print("b={}".format(b))
        
        resnum=1
        for word in W:
            resnum*=m[word-1][s]
            resnum*=spam
        #print("resnumW={}".format(resnum))
        
        cont=0
        for word in V:
            #print("word-1={}".format(word-1))
            if len(b)==3:
                if cont<=0:
                    #print("b[2+cont]={}".format(b[2+cont]))
                    resnum*=m[word-1][s-b[2+cont]]
                else:
                    resnum*=m[word-1][s-1]
                #print("resnumlen3={}".format(resnum))
            elif len(b)==4:
                if cont<=1:
                    resnum*=m[word-1][s-b[2+cont]]
                else:
                    resnum*=m[word-1][s-1]
                #print("resnumlen4={}".format(resnum))
            elif len(b)==5:
                if cont<=2:
                    resnum*=m[word-1][s-b[2+cont]]
                else:
                    resnum*=m[word-1][s-1]
                #print("resnumlen5={}".format(resnum))
            elif len(b)==6:
                if cont<=3:
                    resnum*=m[word-1][s-b[2+cont]]
                else:
                    resnum*=m[word-1][s-1]
                #print("resnumlen6={}".format(resnum))
            cont+=1
        resultadonum+=resnum
        #print("resultadonum={}".format(resultadonum))
    
    V.append(6)
    for a in range(2**len(V)): #Sacamos el denominador
        b=bin(a)
        d=list(b)
        for c in range(len(d)):
            if d[c]=="0":
                d[c]="1"
            else:
                d[c]="0"
        b=list(d)
        for num in range(len(b)):
            b[num]=int(b[num])
        #print("b={}".format(b))
        
        resden=1
        for word in W:
            resden*=m[word-1][s]
        #print("resdenW={}".format(resden))
        
        cont=0
        for word in V:
            #print("word-1={}".format(word-1))
            if len(b)==3:
                if cont<=0:
                    #print("b[2+cont]={}".format(b[2+cont]))
                    resden*=m[word-1][s-b[2+cont]]
                else:
                    resden*=m[word-1][s-1]
                #print("resdenlen3={}".format(resden))
            elif len(b)==4:
                if cont<=1:
                    resden*=m[word-1][s-b[2+cont]]
                else:
                    resden*=m[word-1][s-1]
                #print("resdenlen4={}".format(resden))
            elif len(b)==5:
                if cont<=2:
                    resden*=m[word-1][s-b[2+cont]]
                else:
                    resden*=m[word-1][s-1]
                #print("resdenlen5={}".format(resden))
            elif len(b)==6:
                if cont<=3:
                    resden*=m[word-1][s-b[2+cont]]
                else:
                    resden*=m[word-1][s-1]
                #print("resdenlen6={}".format(resden))
            elif len(b)==7:
                if cont<=4:
                    resden*=m[word-1][s-b[2+cont]]
                else:
                    resden*=m[word-1][s-1]
                #print("resdenlen7={}".format(resden))
            cont+=1
        resultadoden+=resden
        #print("resultadoden={}".format(resultadoden))
            
    res=resultadonum/resultadoden
    #print("Terminé una con spam={} y res={}".format(spam,res))
    V.remove(6)
    return res
```
